```python
x = []
y = []
u = []
v = []
theta = []
p = []
t = []

line_num = 0
R = 0.5
X_min = -4.0
X_max = 12.0
Y_min = -4.0
Y_max = 4.0

N_file = 100
DATA_PATH = "data/"

xyz_file = DATA_PATH + "uvtp_data_" + str(0) + ".csv"

with open(xyz_file, mode='r') as file:
    csvFile = csv.reader(file)
    next(csvFile, None)
    for lines in csvFile:
        x.append(float(lines[3]))
        y.append(float(lines[4]))

for i in range(0, N_file):
    file_name = DATA_PATH + "/uvtp_data_" + str(i) + ".csv"
    time = 0.2 * i
    t.append(time)
    with open(file_name, mode='r') as file:
        csvFile = csv.reader(file)
        next(csvFile, None)
        for lines in csvFile:
            u.append(float(lines[8]))
            v.append(float(lines[9]))
            theta.append(float(lines[7]))
            p.append(float(lines[2]))
```

```python
x = np.array(x)
y = np.array(y)
u = np.array(u)
u = u.reshape(N_file, x.shape[0])
v = np.array(v)
v = v.reshape(N_file, x.shape[0])
theta = np.array(theta)
theta = theta.reshape(N_file, x.shape[0])
p = np.array(p)
p = p.reshape(N_file, x.shape[0])
t = np.array(t)
id_mask = []
for i in range(np.shape(x)[0]):
    if ((x[i] >= X_min) and (x[i] <= X_max)) and ((y[i] >= Y_min) and (y[i] <= Y_max)):
        id_mask.append(i)

x_sub = x[id_mask]
y_sub = y[id_mask]

u_sub = u[:, id_mask]
v_sub = v[:, id_mask]
p_sub = p[:, id_mask]
theta_sub = theta[:, id_mask]

data = {"x": x_sub, "y": y_sub, "t": t,
        "u": u_sub, "v": v_sub, "theta": theta_sub,
        "p": p_sub}

sio.savemat("data.mat", data)
```

```python
# Sample residual points

ub = np.array([np.max(data['x']), np.max(data['y'])])
lb = np.array([np.min(data['x']), np.min(data['y'])])

coord_res = []

for time_step in data['t']:
    sampled_spaces = lb + (ub - lb) * lhs(2, N_r)
    times = time_step * np.ones(N_r) [:, None]
    coord_tmp = np.hstack([times, sampled_spaces])
    coord_res.append(coord_tmp)

coord_res = np.vstack(coord_res)
```

```python
def add_noise(data, level):
    """Add Gaussian noise to data."""
    noise = np.random.normal(0, level, data.shape)
    return data + noise

# Define the noise level
noise_level = 0.01  # 1% noise
```

```python
# Sample initial, boundary and calibration points
# Add some noise to data

id_i = np.random.choice(coord_i_train.shape[0], N_i, replace=False)
coord_i_train = coord_i_train[id_i, :]
momentum_i_train = momentum_i_train[id_i, :]
momentum_i_train = add_noise(momentum_i_train, noise_level)

id_b = np.random.choice(coord_b_train.shape[0], N_b, replace=False)
coord_b_train = coord_b_train[id_b, :]
momentum_b_train = momentum_b_train[id_b, :]
momentum_b_train = add_noise(momentum_b_train, noise_level)

id_p = np.random.choice(coord_p_calib.shape[0], C_p, replace=False)
coord_p_calib = coord_p_calib[id_p, :]
pressure_calib = pressure_calib[id_p, :]
pressure_calib = add_noise(pressure_calib, noise_level)
```

Model Architecture

```python
class ResNetPINN(nn.Module):
    def __init__(self, input_size, output_size, num_neurons, num_blocks, dropout_rate):
        super(ResNetPINN, self).__init__()
        # Input layer
        self.initial_fc = nn.Linear(input_size, num_neurons)
        nn.init.xavier_normal_(self.initial_fc.weight)  

        # Blocks of layers with skip connections
        self.blocks = nn.ModuleList()
        for _ in range(num_blocks):
            block = nn.Sequential(
                nn.Linear(num_neurons, num_neurons),
                nn.BatchNorm1d(num_neurons), 
                nn.Tanh(),
                nn.Dropout(dropout_rate),
            )
            
            nn.init.xavier_normal_(block[0].weight)
            self.blocks.append(block)
        
        # Output layer
        self.output_fc = nn.Linear(num_neurons, output_size)
        nn.init.xavier_normal_(self.output_fc.weight) 

        # Learnable parameter
        self.nu = nn.Parameter(torch.tensor([0.0], requires_grad=True))

    def forward(self, x):
        x = self.initial_fc(x)
        for block in self.blocks:
            residual = x
            x = block(x)
            x += residual  # Implementing the skip connection
        x = self.output_fc(x)
        return x
```

```python
def l2_regularization(model, scale=0.001):
    l2_reg = torch.tensor(0.).to(device)
    for param in model.parameters():
        l2_reg += torch.norm(param)**2
    return scale * l2_reg
    
```

```python
def compute_navier_stokes_residuals(model, coord_res):
    """ Compute residuals for Navier-Stokes equations using model predictions. """
    coord_res.requires_grad_(True)
    outputs = model(coord_res)
    p, u, v = outputs[:, 0], outputs[:, 1], outputs[:, 2]
    
    ones_p = torch.ones_like(p, device=p.device)
    ones_u = torch.ones_like(u, device=u.device)
    ones_v = torch.ones_like(v, device=v.device)
    
    p_grad = grad(p, coord_res, grad_outputs=ones_p, create_graph=True)[0]
    u_grad = grad(u, coord_res, grad_outputs=ones_u, create_graph=True)[0]
    v_grad = grad(v, coord_res, grad_outputs=ones_v, create_graph=True)[0]

    u_t, u_x, u_y = u_grad[:, 0], u_grad[:, 1], u_grad[:, 2]
    v_t, v_x, v_y = v_grad[:, 0], v_grad[:, 1], v_grad[:, 2]
    p_x, p_y = p_grad[:, 1], p_grad[:, 2]

    u_xx = grad(u_x, coord_res, grad_outputs=ones_u, create_graph=True)[0][:, 1]
    u_yy = grad(u_y, coord_res, grad_outputs=ones_u, create_graph=True)[0][:, 2]
    v_xx = grad(v_x, coord_res, grad_outputs=ones_v, create_graph=True)[0][:, 1]
    v_yy = grad(v_y, coord_res, grad_outputs=ones_v, create_graph=True)[0][:, 2]

    # Continuity equation residual (incompressibility)
    continuity_res = u_x + v_y

    # Momentum equations with viscosity term
    momentum_u_res = u_t + u * u_x + v * u_y + p_x - model.nu * (u_xx + u_yy)
    momentum_v_res = v_t + u * v_x + v * v_y + p_y - model.nu * (v_xx + v_yy)

    return continuity_res, momentum_u_res, momentum_v_res
    
```

```python
def compute_loss(model, coord_res, coord_train, momentum, pressure):
    # Physics loss
    continuity_res, momentum_u_res, momentum_v_res = compute_navier_stokes_residuals(model, coord_res)
    phys_loss = torch.mean(continuity_res**2 + momentum_u_res**2 + momentum_v_res**2)

    # Data loss
    predictions = model(coord_train)
    batch_predictions = predictions[:-1]  
    calib_prediction = predictions[-1]  
    u_pred, v_pred = batch_predictions[:, 1], batch_predictions[:, 2]
    p_pred = calib_prediction[0] 
    p_pred = p_pred.unsqueeze(0)
    pressure = pressure.view(1)
    data_loss = F.mse_loss(torch.stack([u_pred, v_pred], dim=1), momentum) + F.mse_loss(p_pred, pressure)

    # # Regularization
    # reg_loss = l2_regularization(model)
    
    loss = phys_loss + data_loss
    
    return loss
```

Hyperparameters and Devices

```python
input_size = 3
output_size = 3
num_neurons = 32
num_blocks = 10
dropout_rate = 0.01

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
coord_res = torch.from_numpy(coord_res).float().to(device)

coord_b_train = torch.from_numpy(coord_b_train).float().to(device)
momentum_b_train = torch.from_numpy(momentum_b_train).float().to(device)
coord_i_train = torch.from_numpy(coord_i_train).float().to(device)
momentum_i_train = torch.from_numpy(momentum_i_train).float().to(device)
coord_p_calib = torch.from_numpy(coord_p_calib).float().to(device)
pressure_calib = torch.from_numpy(pressure_calib).float().to(device)

# Creating DataLoader
coord_train = torch.cat([coord_b_train, coord_i_train], dim=0)
momentum_train = torch.cat([momentum_b_train, momentum_i_train], dim=0)

res_set = TensorDataset(coord_res)
res_loader = DataLoader(res_set, batch_size=200000, shuffle=True)
data_set = TensorDataset(coord_train, momentum_train)
train_loader = DataLoader(data_set, batch_size=N, shuffle=False)
```

```python
def save_checkpoint(model, optimizer, scheduler, optimizer_type, epoch, loss, checkpoint_dir, is_history):
    if is_history:
        filename = f"{optimizer_type}_{epoch}.pth"
    else:
        filename = 'checkpoint.pth'
    file_path = os.path.join(checkpoint_dir, filename)
    checkpoint = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': loss,
        'optimizer_type': optimizer_type
    }
    if scheduler:
        checkpoint['scheduler_state_dict'] = scheduler.state_dict()
    torch.save(checkpoint, file_path)
    
```

```python
def train_model(optimizer, optimizer_type, epochs, model, res_loader, train_loader, coord_p_calib, pressure_calib, compute_loss, checkpoint_dir, checkpoint_path, log_path, device):
    # torch.autograd.set_detect_anomaly(True)

    scheduler = None
    if optimizer_type == 'AdamW':
        scheduler = CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-5)
        
    start_epoch = 0

    with open(log_path, 'a') as log_file:
        if start_epoch == 0: 
            log_file.write("Epoch, Loss, Nu\n")
        
    for epoch in range(start_epoch, epochs):
        
        model.train() 
        total_loss = 0

        selected_indices = random.sample(range(len(res_loader)), 3)
        res_samples = [batch[0].to(device) for index, batch in enumerate(res_loader) if index in selected_indices]
        
        for coord_data, momentum in train_loader:
            coord_data, momentum = coord_data.to(device), momentum.to(device)
            coord_train = torch.cat((coord_data, coord_p_calib), dim=0)

            for coord_res in res_samples:
    
                def closure():
                    optimizer.zero_grad()
                    loss = compute_loss(model, coord_res, coord_train, momentum, pressure_calib) 
                    loss.backward()
                    # torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0) 
                    return loss
        
                loss = optimizer.step(closure)  
                total_loss += loss.item() 
                
        avg_loss = total_loss / len(selected_indices)
        nu_value = model.nu.item()

        with open(log_path, 'a') as log_file:
            log_file.write(f"{epoch}, {avg_loss}, {nu_value}\n")
        
        if epoch % 1000 == 0: 
            save_checkpoint(model, optimizer, scheduler, optimizer_type, epoch, avg_loss, checkpoint_dir, is_history=False)
            save_checkpoint(model, optimizer, scheduler, optimizer_type, epoch, avg_loss, checkpoint_dir, is_history=True)
            print(f'Epoch {epoch}: Loss = {avg_loss}, Nu = {nu_value}')

        if scheduler:
            scheduler.step()
    
    print("Training completed.")
```

Model Training

```python
N_iter_adamw = 30000

adamw_optimizer = AdamW(
    model.parameters(),
    lr=0.002, # 1e-3 ~ 5e-5
)


checkpoint_dir = 'CheckPoints'

model = ResNetPINN(input_size, output_size, num_neurons, num_blocks, dropout_rate).to(device)

if not os.path.exists(checkpoint_dir):
        os.makedirs(checkpoint_dir)
    checkpoint_path = os.path.join(checkpoint_dir, 'checkpoint.pth')
    log_path = os.path.join(checkpoint_dir, 'log.txt') 

if os.path.exists(checkpoint_path):
        checkpoint = torch.load(checkpoint_path)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        if 'scheduler_state_dict' in checkpoint and scheduler:
            scheduler.load_state_dict(checkpoint['scheduler_state_dict'])
        start_epoch = checkpoint['epoch'] + 1
        print(f"Resuming training from epoch {start_epoch}")

# Initial training with AdamW
train_model(adamw_optimizer, 'AdamW', N_iter_adamw, model, res_loader, train_loader, coord_p_calib, pressure_calib, compute_loss, checkpoint_dir, checkpoint_path, log_path, device)

torch.save({
    'epoch': N_iter_adamw,
    'model_state_dict': model.state_dict(),
}, os.path.join(checkpoint_dir, 'model_adamw.pth'))

model = ResNetPINN(input_size, output_size, num_neurons, num_blocks, dropout_rate).to(device)
model_checkpoint = torch.load(os.path.join(checkpoint_dir, 'model_adamw.pth'))
model.load_state_dict(model_checkpoint['model_state_dict'])

N_iter_lbfgs = 20000

lbfgs_optimizer = LBFGS(
    model.parameters(), 
)

# Switch to L-BFGS
train_model(lbfgs_optimizer, 'LBFGS', N_iter_adamw + N_iter_lbfgs, model, res_loader, train_loader, coord_p_calib, pressure_calib, compute_loss, checkpoint_dir, checkpoint_path, log_path, device)
```
